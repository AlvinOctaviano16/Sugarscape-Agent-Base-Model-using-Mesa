#Model Classes

```python
class SugarscapeG1mt(mesa.Model):
  '''
  Manage class to run Sugarscape with Traders
  '''

  def __init__(self,width=15,height=15,torus=False, initial_population=200,min_endowment=25
               ,max_endowment=50,min_metabolism=1,max_metabolism=5,min_vision=1,max_vision=5):
    super().__init__()
    self.width=width
    self.height=height
    self.torus=torus
    self.initial_population=initial_population
    self.min_endowment=min_endowment
    self.max_endowment=max_endowment
    self.min_metabolism=min_metabolism
    self.max_metabolism=max_metabolism
    self.min_vision=min_vision
    self.max_vision=max_vision

    self.grid=mesa.space.MultiGrid(self.width,self.height,self.torus)
    sugar_distribution=np.genfromtxt("sugar-map.txt")
    spice_distribution=np.flip(sugar_distribution,1)

    # agent_id=0 PADA MESA V3 INI TIDAK DIPERLUKAN
    for _,(x,y) in self.grid.coord_iter():
      max_sugar=sugar_distribution[x,y]
      if max_sugar>0:
        '''Perubahan dilakukan pada Spice(agen_id => self) menyesuaikan mesa v3'''
        sugar=Sugar(self,(x,y),max_sugar)
        self.grid.place_agent(sugar,(x,y))
        # agent_id+=1
        # print(sugar)
        # print(sugar.unique_id)

    for x,(a,b) in self.grid.coord_iter():
      max_spice=spice_distribution[a,b]
      if max_spice>0:
        '''Perubahan dilakukan pada Spice(agen_id => self) menyesuaikan mesa v3'''
        spice=Spice(self,(a,b),max_spice)
        self.grid.place_agent(spice,(a,b))
        # agent_id+=1
        # print(spice.unique_id,spice.pos,spice.max_spice)
        print(spice.unique_id)

    for i in range (self.initial_population):
      x_pos=self.random.randrange(self.width)
      y_pos=self.random.randrange(self.height)
      sugar=int(self.random.uniform(min_endowment,max_endowment))
      spice=int(self.random.uniform(min_endowment,max_endowment))
      sugar_metabolism=int(self.random.uniform(min_metabolism,max_metabolism))
      spice_metabolism=int(self.random.uniform(min_metabolism,max_metabolism))
      vision=int(self.random.uniform(min_vision,max_vision))

      trader=Trader(self,
                    (x_pos,y_pos),
                    sugar=sugar,
                    spice=spice,
                    sugar_metabolism=sugar_metabolism,
                    spice_metabolism=spice_metabolism,
                    vision=vision)

      self.grid.place_agent(trader,(x_pos,y_pos))
      # print(trader.unique_id, trader.pos, trader.sugar,trader.spice,trader.sugar_metabolism,trader.spice_metabolism,trader.vision)
      # agent_id+=1


  def step(self):
    self.agents.select(lambda a: type(a) is Sugar).shuffle_do("step")
    self.agents.select(lambda a: type(a) is Spice).shuffle_do("step")
```

```python
class Trader(mesa.Agent):

  '''
  - Setelah membaca dokumentasi saya melakukan perubahan pass parameter pada constructor hal ini dikarenakan , pada mesa.Agent class sudah melakukan assign
  pos dan unique id secara manual. Oleh karena itu saya menghapus unique_id dan pos pada constructor ini dan juga attribut assignnya
  '''

  def __init__(self,model,moore=False,sugar=0,spice=0,spice_metabolism=0,sugar_metabolism=0,vision=0):
    super().__init__(model)
    # self.unique_id=unique_id PADA MESA V3 INI TIDAK DIPERLUKAN KARENA SUDAH DI ASSIGN SECARA OTOMATIS OLEH LIBRARY
    # self.pos=pos PADA MESA V3 INI TIDAK DIPERLUKAN KARENA SUDAH DI ASSIGN SECARA OTOMATIS OLEH LIBRARY
    self.moore=moore
    self.sugar=sugar
    self.spice=spice
    self.spice_metabolism=spice_metabolism
    self.sugar_metabolism=sugar_metabolism
    self.vision=vision

  def get_spice(self,pos):
    '''
    move() helper function for get_amount_spice
    '''
    this_cell=self.model.grid.get_cell_list_contents(pos)
    for agent in this_cell:
      if type(agent) is Spice:
        return agent
    return None


  def get_amount_spice(self,pos):
    '''
    move() helper function for calculate_welfare()
    '''
    spice_patch=self.get_spice(pos)
    if spice_patch:
      return spice_patch.amount
    return 0

  def get_sugar(self,pos):
    '''
    move() helper function for get_amount_sugar()
    '''
    this_cell=self.model.grid.get_cell_list_contents(pos)
    for agent in this_cell:
      if type(agent) is Sugar:
        return agent
    return None

  def get_amount_sugar(self,pos):
    '''
    move() helper function for calculate_welfare()
    '''
    sugar_patch=self.get_sugar(pos)
    if sugar_patch:
      return sugar_patch.amount
    return 0

  def is_occupied_by_other(self,pos):
    '''
    move() helper function
    '''
    if pos==self.pos:
      return False

    this_cell=self.model.grid.get_cell_list_contents(pos)
    for agent in this_cell:
      if isinstance(agent, Trader):
        return True
    return False

  def calculate_welfare(self,sugar,spice):
    '''
    move() helper function
    '''
    m_total=self.sugar_metabolism+self.spice_metabolism
    return sugar**(self.sugar_metabolism/m_total) * spice**(self.spice_metabolism/m_total)

  def move(self):

    '''
    Function for trader agent to identify optimal move for each step in 4 parts
    1 - identify all possible moves
    2 - determine which move maximizes welfare
    3 - find closest best option
    4 - move
    '''

    # 1.identify all possible moves

    neighbours=[i
                for i in self.model.grid.get_neighborhood(
                    self.pos,self.moore,True,self.vision
                )if not self.is_occupied_by_other(i)
    ]

    welfare=[
        self.calculate_welfare(
            self.sugar+self.get_amount_sugar(pos),
            self.spice+self.get_amount_spice(pos)
        ) for pos in neighbours
    ]

    print(self.pos,self.vision,welfare)
```

```python
class Trader(mesa.Agent):

  '''
  - Setelah membaca dokumentasi saya melakukan perubahan pass parameter pada constructor hal ini dikarenakan , pada mesa.Agent class sudah melakukan assign
  pos dan unique id secara manual. Oleh karena itu saya menghapus unique_id dan pos pada constructor ini dan juga attribut assignnya
  '''

  def __init__(self,model,moore=False,sugar=0,spice=0,spice_metabolism=0,sugar_metabolism=0,vision=0):
    super().__init__(model)
    # self.unique_id=unique_id PADA MESA V3 INI TIDAK DIPERLUKAN KARENA SUDAH DI ASSIGN SECARA OTOMATIS OLEH LIBRARY
    # self.pos=pos PADA MESA V3 INI TIDAK DIPERLUKAN KARENA SUDAH DI ASSIGN SECARA OTOMATIS OLEH LIBRARY
    self.moore=moore
    self.sugar=sugar
    self.spice=spice
    self.spice_metabolism=spice_metabolism
    self.sugar_metabolism=sugar_metabolism
    self.vision=vision

  def get_spice(self,pos):
    '''
    move() helper function for get_amount_spice
    '''
    this_cell=self.model.grid.get_cell_list_contents(pos)
    for agent in this_cell:
      if type(agent) is Spice:
        return agent
    return None


  def get_amount_spice(self,pos):
    '''
    move() helper function for calculate_welfare()
    '''
    spice_patch=self.get_spice(pos)
    if spice_patch:
      return spice_patch.amount
    return 0

  def get_sugar(self,pos):
    '''
    move() helper function for get_amount_sugar()
    '''
    this_cell=self.model.grid.get_cell_list_contents(pos)
    for agent in this_cell:
      if type(agent) is Sugar:
        return agent
    return None

  def get_amount_sugar(self,pos):
    '''
    move() helper function for calculate_welfare()
    '''
    sugar_patch=self.get_sugar(pos)
    if sugar_patch:
      return sugar_patch.amount
    return 0

  def is_occupied_by_other(self,pos):
    '''
    move() helper function
    '''
    if pos==self.pos:
      return False

    this_cell=self.model.grid.get_cell_list_contents(pos)
    for agent in this_cell:
      if isinstance(agent, Trader):
        return True
    return False

  def calculate_welfare(self,sugar,spice):
    '''
    move() helper function
    '''
    m_total=self.sugar_metabolism+self.spice_metabolism
    return sugar**(self.sugar_metabolism/m_total) * spice**(self.spice_metabolism/m_total)

  def move(self):

    '''
    Function for trader agent to identify optimal move for each step in 4 parts
    1 - identify all possible moves
    2 - determine which move maximizes welfare
    3 - find closest best option
    4 - move
    '''
    print(f"Posisi saat ini : {self.pos}")

    # 1.identify all possible moves

    neighbours=[i
                for i in self.model.grid.get_neighborhood(
                    self.pos,self.moore,True,self.vision
                )if not self.is_occupied_by_other(i)
    ]

    print(f"Semua tetangga berdasarkan vision : {self.vision} adalah \n {neighbours}")

    # 2. calculating welfare
    welfare=[
        self.calculate_welfare(
            self.sugar+self.get_amount_sugar(pos),
            self.spice+self.get_amount_spice(pos)
        ) for pos in neighbours
    ]

    print(f"Semua nilai welfare berdasarkan tiap tetangga adalah \n {welfare}")

    # 3. a. maximizes welfare
    max_welfare=max(welfare)

    print(f"Nilai maximum welfare : {max_welfare}")

    candidate_indices=[
        i for i in range(len(welfare)) if math.isclose(welfare[i],max_welfare)
    ]

    candidates=[
        neighbours[i] for i in candidate_indices
    ]
    print(f"Semua tetangga yang memiliki maximum welfare adalah \n {candidates}")

    # 3. b. closest best option
    min_dist=min(
        get_distances(self.pos,pos) for pos in candidates
    )

    final_candidates=[
        pos for pos in candidates if math.isclose(get_distances(self.pos,pos),min_dist,rel_tol=1e-02)
    ]

    self.random.shuffle(final_candidates)

    print(f"Ini semua adalah pilihan terbaik untuk move : \n {final_candidates}")

    # 4. move trader position to the closest and welfare pos

    self.model.grid.move_agent(self,final_candidates[0])
    print(f"Posisi Agent saat ini adalah : {self.pos}")

```

#Model Classes

```python
class SugarscapeG1mt(mesa.Model):
  '''
  Manage class to run Sugarscape with Traders
  '''

  def __init__(self,width=5,height=5,torus=False, initial_population=1,min_endowment=25
               ,max_endowment=50,min_metabolism=1,max_metabolism=5,min_vision=1,max_vision=5):
    super().__init__()
    self.width=width
    self.height=height
    self.torus=torus
    self.initial_population=initial_population
    self.min_endowment=min_endowment
    self.max_endowment=max_endowment
    self.min_metabolism=min_metabolism
    self.max_metabolism=max_metabolism
    self.min_vision=min_vision
    self.max_vision=max_vision

    self.grid=mesa.space.MultiGrid(self.width,self.height,self.torus)
    sugar_distribution=np.genfromtxt("/content/drive/MyDrive/sugar-map-2.txt")
    spice_distribution=np.flip(sugar_distribution,1)

    # agent_id=0 PADA MESA V3 INI TIDAK DIPERLUKAN
    for _,(x,y) in self.grid.coord_iter():
      max_sugar=sugar_distribution[y,x]
      if max_sugar>0:
        '''Perubahan dilakukan pada Spice(agen_id => self) menyesuaikan mesa v3'''
        sugar=Sugar(self,max_sugar)
        self.grid.place_agent(sugar,(x,y))
        # agent_id+=1
        # print(sugar)
        # print(sugar.unique_id)

    for x,(a,b) in self.grid.coord_iter():
      max_spice=spice_distribution[b,a]
      if max_spice>0:
        '''Perubahan dilakukan pada Spice(agen_id => self) menyesuaikan mesa v3'''
        spice=Spice(self,max_spice)
        self.grid.place_agent(spice,(a,b))
        # agent_id+=1
        # print(spice.unique_id,spice.pos,spice.max_spice)
        # print(spice.unique_id)

    for i in range (self.initial_population):
      x_pos=self.random.randrange(self.width)
      y_pos=self.random.randrange(self.height)
      sugar=int(self.random.uniform(min_endowment,max_endowment))
      spice=int(self.random.uniform(min_endowment,max_endowment))
      sugar_metabolism=int(self.random.uniform(min_metabolism,max_metabolism))
      spice_metabolism=int(self.random.uniform(min_metabolism,max_metabolism))
      vision=int(self.random.uniform(min_vision,max_vision))

      trader=Trader(self,
                    # (x_pos,y_pos),
                    sugar=sugar,
                    spice=spice,
                    sugar_metabolism=sugar_metabolism,
                    spice_metabolism=spice_metabolism,
                    vision=vision)

      self.grid.place_agent(trader,(x_pos,y_pos))
      # print(trader.unique_id, trader.pos, trader.sugar,trader.spice,trader.sugar_metabolism,trader.spice_metabolism,trader.vision)
      # agent_id+=1


  def step(self):
    # Resources Step
    self.agents.select(lambda a: type(a) is Sugar).shuffle_do("step")
    self.agents.select(lambda a: type(a) is Spice).shuffle_do("step")
    # Trader Move
    self.agents.select(lambda a: type(a) is Trader).shuffle_do("move")
    self.steps+=1


  # def step(self):
  #   '''Mesa v2'''
  #   for sugar in self.schedule.agents_by_type[Sugar].values():
  #     sugar.step()


  def run_model(self,step_count=100):
    for i in range(step_count):
      self.step()
```

#Helper Function

```python
def get_distances(pos1,pos2):
  '''
  This is a helper function, to get Euclidean distance between two position
  used in trader.move()
  '''
  x1,y1=pos1
  x2,y2=pos2
  dx=x1-x2
  dy=y1-y2
  return math.sqrt(dx**2 + dy**2)
```

```python
class Sugar(mesa.Agent):
    """
    Sugar:
    - contains an amount of sugar
    - grows 1 amount of sugar at each turn
    """

    '''
    - Setelah membaca dokumentasi saya melakukan perubahan pass parameter pada constructor hal ini dikarenakan , pada mesa.Agent class sudah melakukan assign
    pos dan unique id secara manual. Oleh karena itu saya menghapus unique_id dan pos pada constructor ini dan juga attribut assignnya
    '''
    def __init__(self, model, max_sugar):
        super().__init__(model)

        # self.unique_id = unique_id
        self.amount = max_sugar
        self.max_sugar = max_sugar

    def step(self):
      '''
      Sugar growth function, adds one units untill max_sugar is reached
      '''
      self.amount=min([self.amount+1,self.max_sugar])
      # print(self.unique_id,self.amount,self.max_sugar)
```
